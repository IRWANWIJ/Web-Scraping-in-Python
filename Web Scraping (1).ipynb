```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://id.wikipedia.org/wiki/Daftar_perusahaan_tertua_di_Indonesia'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
column_titles = table.find_all('th')
```

```python
print(column_titles)
```

```python
column_table_titles = [title.text.strip() for title in column_titles]

print (column_table_titles)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns = column_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data :
    row_data = row.find_all('td')
   
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data:
    row_data = row.find_all('td')
   
    
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
#[], index = 0
#['1878', 'Koffie Warung Tinggi', 'Jakarta', 'Kedai kopi', '[7]'], index = 7
#notice above data, for data in index 0, has no data, 
#for data in index 7, only has 5 column instead of 6
#therefore we use if and continue to skip index 0 and index 7


for row in column_data :
    row_data = row.find_all('td')
    if len(row_data) != 6 :
        continue
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data:
    row_data = row.find_all('td')
    if len(row_data) != 6 :
        continue
    
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
